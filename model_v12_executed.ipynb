```python
import copy

ha = sj.create_model([
    sdf_ss_D, government_default_D, bond_price_ss_D, bond_return_D,
    sdf_ss_F, government_default_F, bond_price_ss_F, bond_return_F,
    hh_extended_D, smart_steady_D, market_clearing_D, steady_auxilliary_D,
    banker_div_D, government_ss_D, labor_ss_D,
    hh_extended_F, smart_steady_F, market_clearing_F, steady_auxilliary_F,
    banker_div_F, government_ss_F, labor_ss_F,
    ces_price_D, import_demand_D, ces_price_F, import_demand_F,
    deposit_return_D, deposit_return_F,
    bond_yield,
    trade_balance, external_account_D, global_goods_mkt,
], name='MU HA Model 2 Country')

unknowns_ss = {'beta_D': 0.9678948189, 'beta_F': 0.9654079110, 'p': 0.997170}
targets_ss  = ['deposit_mkt_D', 'deposit_mkt_F', 'ca_res_D']

ss = ha.solve_steady_state(calibration_start, unknowns_ss, targets_ss, solver='broyden_custom')

# ── Post-solve anchors ────────────────────────────────────────────────────────
anchors = {
    # Share-based values for domestic bond FOCs (domestic_bond_foc_D/F)
    'phi_bD_D_ss':           float(ss['q_b_D']) * float(ss['b_D_D']) / float(ss['n_inter_D']),
    'phi_bF_F_ss':           float(ss['q_b_F']) * float(ss['b_F_F']) / (float(ss['p']) * float(ss['n_inter_F'])),
    # Level anchors → portfolio_level_anchors block → b_F_D_ss, b_D_F_ss
    'b_F_D_anchor':          float(ss['b_F_D']),
    'b_D_F_anchor':          float(ss['b_D_F']),
    # Excess returns used by domestic_bond_foc_D/F and portfolio_adj_cost
    'excess_return_bD_D_ss': float(ss['rb_actual_D']) - float(ss['rdep_D']) - calibration_start['T0_D'],
    'excess_return_bF_F_ss': float(ss['rb_actual_F']) - float(ss['rdep_F']) - calibration_start['T0_F'],
    'excess_return_F_D_ss':  float(ss['rb_actual_F']) - float(ss['rdep_D']) - calibration_start['T0_D'],
    'excess_return_D_F_ss':  float(ss['rb_actual_D']) - float(ss['rdep_F']) - calibration_start['T0_F'],
    'q_b_D':                 float(ss['q_b_D']),
    'q_b_F':                 float(ss['q_b_F']),
    'p':                     float(ss['p']),
}
calibration_start.update(anchors)
for k, v in anchors.items():
    ss.toplevel[k] = v

ss.toplevel['b_F_D_ss'] = float(ss['b_F_D'])
ss.toplevel['b_D_F_ss'] = float(ss['b_D_F'])


ss.toplevel['Rgross_D'] = float(1 + ss['rdep_D'])
ss.toplevel['Rgross_F'] = float(1 + ss['rdep_F'])

_fr_D = float(ss['frisch_D']); _fr_F = float(ss['frisch_F'])
ss.toplevel['X_D'] = float(ss['C_D']) - float(ss['vphi_D']) * float(ss['N_D']) ** (1 + 1/_fr_D) / (1 + 1/_fr_D)
ss.toplevel['X_F'] = float(ss['C_F']) - float(ss['vphi_F']) * float(ss['N_F']) ** (1 + 1/_fr_F) / (1 + 1/_fr_F)

# NOTE: Phi_D / Phi_F overrides REMOVED. smart_steady_{D,F} compute them at
# their true SS value (~1.7e-4 here), and m_{D,F} was set assuming that value.
# Overriding to 0 broke intermediation_P2 and market_clearing in linearisation.
for k, v in {
    'tau_mp_D':        0.0,  'tau_mp_F':        0.0,
    'T_D':             0.0,  'T_F':             0.0,   # T0=T1=0, so T is truly zero
    'b_F_D_res':       0.0,  'b_D_F_res':       0.0,
    'rb_D_res':        0.0,  'rb_F_res':        0.0,
    'labor_mkt_res_D': 0.0,  'labor_mkt_res_F': 0.0,
    'w_res_D':         0.0,  'w_res_F':         0.0,
}.items():
    ss.toplevel[k] = v

# ── SS Bellman values for three ν's (consumed by ha_full as inner unknowns) ──
for k in ('nu_K_D', 'nu_bD_D', 'nu_bF_D', 'eta_D',
          'nu_K_F', 'nu_bF_F', 'nu_bD_F', 'eta_F'):
    if k not in ss.toplevel:
        ss.toplevel[k] = 0.0

cali_D = cali_F = ss
ss_final = copy.deepcopy(ss)

print(f"beta_D={ss['beta_D']:.10f}  beta_F={ss['beta_F']:.10f}  p={ss['p']:.6f}")
print(f"lambda_gk_D={ss['lambda_gk_D']:.6f}  lambda_gk_F={ss['lambda_gk_F']:.6f}")
print(f"rb_D={ss['rb_D']:.6f}  rb_F={ss['rb_F']:.6f}  rdep_D={ss['rdep_D']:.6f}  rdep_F={ss['rdep_F']:.6f}")
print(f"q_b_D={anchors['q_b_D']:.6f}  q_b_F={anchors['q_b_F']:.6f}")
print(f"vphi_D={float(ss['vphi_D']):.6f}  vphi_F={float(ss['vphi_F']):.6f}")
print(f"X_D={ss.toplevel['X_D']:.6f}  X_F={ss.toplevel['X_F']:.6f}")
print(f"Phi_D={float(ss['Phi_D']):.6e}  Phi_F={float(ss['Phi_F']):.6e}  (preserved from smart_steady)")
```

```python
# ── SS residual diagnostic ─────────────────────────────────────────────────────
def _get(k):
    return float(ss[k])

diag = {}

# ── Multi-asset GK IC ──────────────────────────────────────────────────────────
for c in ['D', 'F']:
    pdiv    = _get('p') if c == 'F' else 1.0
    eta_c   = _get(f'eta_{c}')
    lam     = _get(f'lambda_gk_{c}')
    theta_c = _get(f'theta_{c}')
    Q_c     = _get(f'Q_{c}')
    K_c     = _get(f'K_{c}')
    n_c     = _get(f'n_inter_{c}')
    kappa_c = Q_c * K_c / n_c
    if c == 'D':
        nu_K, nu_bD, nu_bF = _get('nu_K_D'), _get('nu_bD_D'), _get('nu_bF_D')
        q_h, q_x = _get('q_b_D'), _get('q_b_F')
        b_h, b_x = _get('b_D_D'), _get('b_F_D')
    else:
        nu_K, nu_bD, nu_bF = _get('nu_K_F'), _get('nu_bD_F'), _get('nu_bF_F')
        q_h, q_x = _get('q_b_F'), _get('q_b_D')
        b_h, b_x = _get('b_F_F'), _get('b_D_F')
    phi_h = q_h * b_h / (pdiv * n_c)
    phi_x = q_x * b_x / (pdiv * n_c)
    LHS = (nu_K * kappa_c
           + (nu_bD if c == 'D' else nu_bF) * phi_h
           + (nu_bF if c == 'D' else nu_bD) * phi_x
           + eta_c)
    diag[f'IC_{c}: LHS − λ·θ'] = LHS - lam * theta_c

# ── Bellman P1 residuals ───────────────────────────────────────────────────────
for c in ['D', 'F']:
    f_c    = _get(f'f_{c}')
    lam    = _get(f'lambda_gk_{c}')
    th     = _get(f'theta_{c}')
    beta_c = _get(f'beta_{c}')
    rk_c   = _get(f'rk_{c}')
    rdep_c = _get(f'rdep_{c}')
    eta_c  = _get(f'eta_{c}')
    Omega_p1 = f_c + (1 - f_c) * lam * th
    rb_h = _get('rb_actual_D' if c == 'D' else 'rb_actual_F')
    rb_x = _get('rb_actual_F' if c == 'D' else 'rb_actual_D')
    nu_K_c  = _get(f'nu_K_{c}')
    nu_bh_c = _get('nu_bD_D' if c == 'D' else 'nu_bF_F')
    nu_bx_c = _get('nu_bF_D' if c == 'D' else 'nu_bD_F')
    diag[f'P1_{c}: nu_K_res']  = nu_K_c  - beta_c * Omega_p1 * (rk_c - rdep_c)
    diag[f'P1_{c}: nu_bh_res'] = nu_bh_c - beta_c * Omega_p1 * (rb_h - rdep_c)
    diag[f'P1_{c}: nu_bx_res'] = nu_bx_c - beta_c * Omega_p1 * (rb_x - rdep_c)
    diag[f'P1_{c}: eta_res']   = eta_c   - beta_c * Omega_p1 * (1 + rdep_c)

# ── GHH labor market FOC: w/P_CES = vphi·N^(1/frisch) (UCE cancels out) ──────
for c in ['D', 'F']:
    w_c    = _get(f'w_{c}')
    P_c    = _get(f'P_CES_{c}')
    N_c    = _get(f'N_{c}')
    vphi_c = _get(f'vphi_{c}')
    fr_c   = _get(f'frisch_{c}')
    diag[f'labor_mkt_{c}: w/P − vphi·N^(1/fr)'] = w_c / P_c - vphi_c * N_c ** (1 / fr_c)

# ── Cross-border portfolio FOC (level-based, matches portfolio_adj_cost) ───────
for (c, cross, b_key, b_ss_key, psi_key, er_key) in [
    ('D', 'F', 'b_F_D', 'b_F_D_ss', 'psi_bF_D', 'excess_return_F_D_ss'),
    ('F', 'D', 'b_D_F', 'b_D_F_ss', 'psi_bD_F', 'excess_return_D_F_ss'),
]:
    diag[f'portfolio_adj_b{cross}_{c}'] = (
        (_get(f'rb_actual_{cross}') - _get(f'rdep_{c}'))
        - _get(er_key)
        - calibration_start[psi_key] * (_get(b_key) - _get(b_ss_key))
        - _get(f'tau_mp_{c}')
    )

# ── Domestic bond FOC (share-based, matches domestic_bond_foc_D/F) ────────────
for c in ['D', 'F']:
    pdiv       = _get('p') if c == 'F' else 1.0
    phi_dom    = _get(f'q_b_{c}') * _get(f'b_{c}_{c}') / (pdiv * _get(f'n_inter_{c}'))
    phi_dom_ss = _get(f'phi_b{c}_{c}_ss')
    psi_dom    = _get(f'psi_b{c}_{c}')
    er_dom     = _get(f'excess_return_b{c}_{c}_ss')
    diag[f'dom_bond_foc_{c}'] = (
        (_get(f'rb_actual_{c}') - _get(f'rdep_{c}'))
        - er_dom
        - psi_dom * (phi_dom - phi_dom_ss)
        - calibration_start[f'T0_{c}']
    )

diag['ca_res_D'] = _get('ca_res_D')

TOL = 1e-8
print(f"\n{'Block residual':<52} {'Value':>14}  Status")
print("-" * 82)
FLAGGED = []
for name, val in diag.items():
    ok = abs(val) <= TOL
    if not ok:
        FLAGGED.append(name)
    print(f"  {name:<50} {val:>14.6e}  {'OK' if ok else '*** FAIL'}")
print("-" * 82)
print("\nAll residuals < 1e-8  ✓" if not FLAGGED else f"\nFLAGGED: {FLAGGED}")
```

```python
BLUE    = '#002147'
RED   = '#8C1515'
BLUE_MUTED     = '#4a6f8a'  
RED_MUTED      = '#c0624a'

_COLORS     = [BLUE, RED, BLUE_MUTED, RED_MUTED]
_LINESTYLES = ['-', '--', '-.', ':']
_MARKERS    = ['', '', '', 'o']

def show_irfs(irfs_list, variables, labels=None,
              ylabel='Deviation from SS (pp)', T_plot=100, figsize=(18, 5)):

    labels = labels or [''] * len(irfs_list)
    n_var  = len(variables)
    fig, axes = plt.subplots(1, n_var, figsize=figsize, sharey=False)
    if n_var == 1:
        axes = [axes]

    for i, (ax, var) in enumerate(zip(axes, variables)):
        for j, (irf, label) in enumerate(zip(irfs_list, labels)):
            data = irf[var][:T_plot] if var in irf else np.zeros(T_plot)
            mkr  = _MARKERS[j % len(_MARKERS)]
            ax.plot(data,
                    color     = _COLORS[j % len(_COLORS)],
                    linestyle = _LINESTYLES[j % len(_LINESTYLES)],
                    linewidth = 1.8,
                    marker    = mkr,
                    markersize= 4,
                    markevery = 4,
                    label     = label)

        ax.axhline(0, color='#888888', linewidth=0.8, linestyle=':')
        ax.set_title(var, fontsize=10, pad=6)
        ax.set_xlabel('Quarter', fontsize=9)
        if i == 0:
            ax.set_ylabel(ylabel, fontsize=9)
        ax.spines[['top', 'right']].set_visible(False)
        ax.tick_params(labelsize=8)
        if any(l for l in labels):
            ax.legend(fontsize=8, frameon=False)

    fig.tight_layout()
    plt.show()
```

```python
# ── WALRAS PROBE: SS baseline + linearised dynamic check ────────────────────
# Probe formulas match the current model:
#   - intermediation_P2 / banker_div_res: NO cap_profit (bank doesn't get it)
#   - income_{D,F}: HH receives cap_profit_{D,F} as uniform lump-sum dividend
#   - market_clearing: NO cap_profit (physical resource constraint)

T_probe = 500
rho_probe = 0.8
shock_probe = 1e-5
dZ_small  = shock_probe * rho_probe ** np.arange(T_probe)
dDef_small = shock_probe * rho_probe ** np.arange(T_probe)

print("Computing tiny IRFs for linearity check...")
irfs_Z_small   = ha_full.solve_impulse_linear(ss_final, unknowns_tp, targets_tp, {'Z_D': dZ_small})
irfs_def_small = ha_full.solve_impulse_linear(ss_final, unknowns_tp, targets_tp, {'shock_def_D': dDef_small})


def _lvl(ss, irf, var, t):
    base = float(ss[var]) if (var in ss.toplevel or var in ss.internals) else 0.0
    if irf is None:
        return base
    if var in irf:
        return base + irf[var][t]
    return base


def _walras_probe(ss, irf_big, irf_small, label, ts=(0, 1, 2, 5, 10, 50)):
    print(f"\n{'═' * 110}\nWALRAS PROBE — {label}\n{'═' * 110}")

    def make_L(irf):
        def L(var, t, lag=0):
            if t - lag < 0:
                return float(ss[var]) if (var in ss.toplevel or var in ss.internals) else 0.0
            return _lvl(ss, irf, var, t - lag)
        return L

    def compute_all(L, ts_local):
        rows = []
        for c in ['D', 'F']:
            d   = float(ss[f'delta_{c}']); ksi = float(ss[f'ksi_{c}'])
            g0  = float(ss[f'gamma0_{c}']); g1  = float(ss[f'gamma1_{c}'])
            fc  = float(ss[f'f_{c}']);   a   = float(ss[f'alpha_{c}'])

            rows.append((f'cap_law_{c}',
                {t: L(f'K_{c}', t) - (1 - d) * L(f'K_{c}', t, 1)
                    - (g0 * (L(f'I_{c}', t) / L(f'K_{c}', t, 1)) ** (1 - ksi) + g1) * L(f'K_{c}', t, 1)
                    for t in ts_local}))
            rows.append((f'cap_profit_def_{c}',
                {t: L(f'cap_profit_{c}', t)
                    - (L(f'Q_{c}', t) * (L(f'K_{c}', t) - (1 - d) * L(f'K_{c}', t, 1))
                       - L(f'I_{c}', t)) for t in ts_local}))

            # market_clearing — matches equations: NO cap_profit
            def gm(t, c=c):
                Y = L(f'Y_{c}', t); C = L(f'C_{c}', t); I = L(f'I_{c}', t)
                G = L(f'G_{c}', t); Phi = L(f'Phi_{c}', t); T = L(f'T_{c}', t)
                NX = L(f'NX_{c}', t); P = L(f'P_CES_{c}', t)
                return Y - (P * C + I + G + Phi + T) - NX
            rows.append((f'goods_mkt_{c}', {t: gm(t) for t in ts_local}))

            # Gov budget — matches budget_residual_D/F (F applies /p)
            def govres_D(t):
                bg = L('b_gov_D', t); bgl = L('b_gov_D', t, 1)
                dr = L('def_rate_D', t); qb = L('q_b_D', t)
                rrD = float(ss['recovery_rate_D']); zwD = float(ss['zeta_writeoff_D'])
                dbD = float(ss['delta_b_D']); hD = 1 - rrD
                sc = 1 - zwD * dr * hD
                coupon = dbD * (1 - dr * hD) * bgl
                net_iss = qb * (bg - sc * (1 - dbD) * bgl)
                return coupon + L('G_D', t) - L('P_CES_D', t) * L('TAX_D', t) - net_iss
            def govres_F(t):
                bg = L('b_gov_F', t); bgl = L('b_gov_F', t, 1)
                dr = L('def_rate_F', t); qb = L('q_b_F', t); p = L('p', t)
                rrF = float(ss['recovery_rate_F']); zwF = float(ss['zeta_writeoff_F'])
                dbF = float(ss['delta_b_F']); hF = 1 - rrF
                sc = 1 - zwF * dr * hF
                coupon = dbF * (1 - dr * hF) * bgl
                net_iss = qb * (bg - sc * (1 - dbF) * bgl)
                return (coupon - net_iss) / p + L('G_F', t) - L('P_CES_F', t) * L('TAX_F', t)
            if c == 'D':
                rows.append((f'gov_budget_D', {t: govres_D(t) for t in ts_local}))
            else:
                rows.append((f'gov_budget_F', {t: govres_F(t) for t in ts_local}))

            # Bank wealth flow — matches intermediation_P2: NO cap_profit
            def bankflow(t, c=c, fc=fc):
                gross = (1 + L(f'rn_{c}', t)) * L(f'n_inter_{c}', t, 1)
                return (1 - fc) * gross + L(f'm_{c}', t) - L(f'Phi_{c}', t) - L(f'T_{c}', t) - L(f'n_inter_{c}', t)
            rows.append((f'bank_flow_{c}', {t: bankflow(t) for t in ts_local}))

            # Bank balance sheet — matches k_balance_sheet (F applies /p)
            def bs_D(t):
                return (L('Q_D', t) * L('K_D', t) + L('q_b_D', t) * L('b_D_D', t)
                        + L('q_b_F', t) * L('b_F_D', t) - L('theta_D', t) * L('n_inter_D', t))
            def bs_F(t):
                p = L('p', t)
                return (L('Q_F', t) * L('K_F', t)
                        + (L('q_b_F', t) * L('b_F_F', t) + L('q_b_D', t) * L('b_D_F', t)) / p
                        - L('theta_F', t) * L('n_inter_F', t))
            if c == 'D':
                rows.append(('bank_bs_D', {t: bs_D(t) for t in ts_local}))
            else:
                rows.append(('bank_bs_F', {t: bs_F(t) for t in ts_local}))

            # Bank dividend — matches banker_div_res: NO cap_profit
            def divres(t, c=c, fc=fc):
                gross = (1 + L(f'rn_{c}', t)) * L(f'n_inter_{c}', t, 1)
                return L(f'div_{c}', t) - (fc * gross - L(f'm_{c}', t))
            rows.append((f'div_def_{c}', {t: divres(t) for t in ts_local}))

            # HH aggregate budget — HH now receives cap_profit_{c} on top of div_{c}
            def hh(t, c=c):
                DEP = L(f'DEP_{c}', t); DEPl = L(f'DEP_{c}', t, 1)
                C = L(f'C_{c}', t); TAX = L(f'TAX_{c}', t)
                w = L(f'w_{c}', t); N = L(f'N_{c}', t)
                div = L(f'div_{c}', t); cp = L(f'cap_profit_{c}', t)
                P = L(f'P_CES_{c}', t); Pl = L(f'P_CES_{c}', t, 1)
                Rg = (1 + L(f'rdep_{c}', t)) * Pl / P
                return (DEP + C) - (Rg * DEPl + (w * N + div + cp) / P - TAX)
            rows.append((f'hh_budget_{c}', {t: hh(t) for t in ts_local}))

            rows.append((f'prod_{c}',
                {t: L(f'Y_{c}', t) - L(f'Z_{c}', t) * L(f'K_{c}', t, 1) ** a * L(f'N_{c}', t) ** (1 - a)
                 for t in ts_local}))

        rows.append(('trade_close',
            {t: L('NX_D', t) + L('p', t) * L('NX_F', t) for t in ts_local}))
        rows.append(('NX_D_def',
            {t: L('NX_D', t) - (L('IM_F', t) - L('p', t) * L('IM_D', t)) for t in ts_local}))
        rows.append(('NX_F_def',
            {t: L('NX_F', t) - (L('IM_D', t) - L('IM_F', t) / L('p', t)) for t in ts_local}))

        def ca(t):
            rcp = (1 + L('rb_actual_F', t)) * L('q_b_F', t, 1) * L('b_F_D', t, 1)
            pay = (1 + L('rb_actual_D', t)) * L('q_b_D', t, 1) * L('b_D_F', t, 1)
            nfa = L('q_b_F', t) * L('b_F_D', t) - L('q_b_D', t) * L('b_D_F', t)
            return L('NX_D', t) + rcp - pay - nfa
        rows.append(('ca_res_D_check', {t: ca(t) for t in ts_local}))
        return rows

    ss_only = compute_all(make_L(None),     [0])
    big     = compute_all(make_L(irf_big),   ts)
    small   = compute_all(make_L(irf_small), ts)

    SS_TOL = 1e-7
    print(f"\n{'Identity':<22} {'SS':>11} {'max|small|':>12} {'max|big|':>11} {'big/small':>11}  Diagnosis")
    print("-" * 100)
    for (name_s, vals_s), (_, vals_sm), (_, vals_b) in zip(ss_only, small, big):
        v_ss = vals_s[0]
        v_sm = max(abs(v) for v in vals_sm.values())
        v_bg = max(abs(v) for v in vals_b.values())
        ratio_str = f"{v_bg/v_sm:>9.1e}" if v_sm > 1e-14 else "       ∞"

        if abs(v_ss) > SS_TOL:
            diag = "*** SS LEAK"
        elif v_sm < 1e-9:
            diag = "clean"
        elif v_sm > 1e-9 and v_bg / max(v_sm, 1e-30) < 1e4:
            diag = "*** LIN LEAK"
        elif v_bg / max(v_sm, 1e-30) > 1e5:
            diag = "quad noise"
        else:
            diag = "mixed"

        print(f"{name_s:<22} {v_ss:>+11.2e} {v_sm:>12.2e} {v_bg:>11.2e} {ratio_str:>11}  {diag}")


_walras_probe(ss_final, irfs_Z_D,   irfs_Z_small,   "TFP shock (1e-2 vs 1e-5)")
_walras_probe(ss_final, irfs_def_D, irfs_def_small, "Default shock (1e-2 vs 1e-5)")
```
